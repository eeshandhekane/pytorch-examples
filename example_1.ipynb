```python
# Dependencies
import torch
```

```python
# Initializing tensors-- zeros
zeros = torch.zeros(2, 3)
print(zeros)
print(zeros.dtype)

zeros_int = torch.zeros(2, 3, dtype=torch.int32)
print(zeros_int)
print(zeros_int.dtype)

zeros_int_another = torch.zeros(2, 3).int()
print(zeros_int_another)
print(zeros_int_another.dtype)
```

```python
# Advanced operations. Certain operations are only on floats.
torch.manual_seed(seed=7)
x = torch.randint(low=0, high=10, size=[2, 3], dtype=torch.float32)

print('min(\n', x, '\n) = ', torch.min(x))
print('*'*79)

print('max(\n', x, '\n) = ', torch.max(x))
print('*'*79)

print('mean(\n', x, '\n) = ', torch.mean(x))
print('*'*79)

print('std(\n', x, '\n) = ', torch.std(x))
print('*'*79)

print('mean_and_std(\n', x, '\n) = ', torch.std_mean(x))  # Returns the std and mean as tensors in a tuple.
print('*'*79)

torch.manual_seed(seed=7)
x = torch.randint(low=0, high=10, size=[2, 3, 4], dtype=torch.float32)

# In case multiple indices for match of min/max, the first index of a match is returned.
print('min_along_dim_0(\n', x, '\n) = ', torch.min(x, dim=0, keepdim=True)[0])
print('min_along_dim_0_indices(\n', x, '\n) = ', torch.min(x, dim=0, keepdim=True)[1])
print('*'*79)

print('max_along_dim_1(\n', x, '\n) = ', torch.max(x, dim=1, keepdim=True)[0])
print('max_along_dim_1_indices(\n', x, '\n) = ', torch.max(x, dim=1, keepdim=True)[1])
print('*'*79)

print('max_along_dim_2(\n', x, '\n) = ', torch.max(x, dim=2, keepdim=True)[0])
print('max_along_dim_2_indices(\n', x, '\n) = ', torch.max(x, dim=2, keepdim=True)[1])
print('*'*79)
```

```python
# Other mathematical operations.

torch.manual_seed(seed=7)
x = torch.randint(low=0, high=10, size=[2, 3]).float()

print('x = ', x)

print('x * 2 = ', x*2)
print('x + 2 = ', x+2)
print('x / 2 = ', x/2)
print('x - 2 = ', x-2)

print('*'*79)

x = torch.rand(2, 3)*2 - 1.0
print('x = ', x)

print('arctan(tan(\n', x, '\n))\n=\n', torch.arctan(torch.tan(x)))
print('*'*79)
print('sin(\n', x, '\n)\n=\n', torch.sin(x))
print('*'*79)
print('arccos(\n', x, '\n)\n=\n', torch.arccos(x))
print('*'*79)
print('absolute_value(\n', x, '\n)\n=\n', torch.abs(x))  # torch.absolute also works
print('*'*79)
```

```python
# Basic linear algebra

torch.manual_seed(seed=7)
x = torch.rand(2, 2)

print('determinant(\n', x, '\n) = ', torch.det(x))
print('*'*79)
print('eigenpair(\n', x, '\n) = ', torch.eig(input=x, eigenvectors=True))
print('*'*79)

torch.manual_seed(seed=7)
x = torch.rand(3, 2)  # SVD of a single matrix
print('svd(\n', x, '\n) = ', torch.linalg.svd(x, full_matrices=False))
u, s, v_transpose = torch.linalg.svd(x, full_matrices=False)
print('x.shape:', x.shape, 'U.shape:', u.shape, 'S.shape:', s.shape, 'V.shape:', v_transpose.shape)
print('U*S*V^T = ', u @ torch.diag_embed(s) @ v_transpose)
print('U*S*V^T = ', torch.mm(torch.mm(u, torch.diag(s)), v_transpose))
print('error(x, U*S*V^T) = ', torch.dist(torch.mm(torch.mm(u, torch.diag(s)), v_transpose), x))
print('*'*79)

torch.manual_seed(seed=7)
x = torch.rand(4, 3, 2)  # SVD of a batch of matrices
print('svd(\n', x, '\n) = ', torch.linalg.svd(x, full_matrices=False))
u, s, v_transpose = torch.linalg.svd(x, full_matrices=False)
print('x.shape:', x.shape, 'U.shape:', u.shape, 'S.shape:', s.shape, 'V.shape:', v_transpose.shape)
print('U*S*V^T = ', u @ torch.diag_embed(s) @ v_transpose)
"""
@ performs ([m, n], [n, p] -> [m, p]) matrix multiplications on [<dims>, m, n]
and [<dims>, n, p] tensors to produces [<dims>, m, p] output.
diag_embed changes a [<dims>, n] dimensional array into [<dims>, n, n] tensor
with nxn diagonal matrices for each of the rows.
"""
print('error(x, U*S*V^T) = ', torch.dist(u @ torch.diag_embed(s) @ v_transpose, x))
print('*'*79)
```
